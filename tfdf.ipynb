```python
import tensorflow_decision_forests as tfdf
import pandas as pd
```

```python
dataset = pd.read_csv("/var/wd_smit/localdata/datasets/Dataset.csv")
```

```python
dataset
```

```python
%pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(dataset, test_size=0.2, random_state=0)
```

```python
train
```

```python
from tensorflow_decision_forests.keras import Task
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train, label="SalePrice", task=Task.REGRESSION)
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test, label="SalePrice", task=Task.REGRESSION)
```

```python
model = tfdf.keras.RandomForestModel(task=Task.REGRESSION)
model.fit(train_ds)
```

```python
model.summary()
```

```python
model.evaluate(test_ds)
```
